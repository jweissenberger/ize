# Line Charts

```python
import pandas as pd
import numpy as np
import altair as alt
```

```python
x = np.arange(100)
df = pd.DataFrame({
  'x': x,
  'f(x)': np.sin(x / 5)
})
df.head()
```

```python
"""
# chage up the names here
plot x vs f(x)
create a line chart of x on the x axis and fx on the y axis
plot a line graph of x and fx
show a line chart of x on the x-axis and fx on the y-axis
"""
```

```python
alt.Chart(df).mark_line().encode(
    x='x',
    y='f(x)'
)
```

```python
"""
create a line chart of x and fx and plot the points
graph a line chart of x and fx and show the datapoints
plot a line graph of x and fx and display the individual data points
"""
```

```python
alt.Chart(df).mark_line(point=True).encode(
    x='x',
    y='f(x)'
)
```

# Bar Charts

```python
df = pd.DataFrame({
    'a': ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I'],
    'b': [28, 55, 43, 91, 81, 53, 19, 87, 52]
})
```

```python
"""
create a bar graph with column a on the x axis and column b on the y axis
make a bar chart of column a against column b
show a bar graph with columns a and b
make a bar graph of a and b
"""
```

```python
alt.Chart(df).mark_bar().encode(
    x='a',
    y='b'
)
```

```python
df = data.us_employment()
```

```python
"""
You want to teach it that if its plotting time on the xaxis you need to put :T and that time always needs to be
on the x-axis

teach it that if they specify time you need to look for month/year/date etc column

Create a bar chart of nonfarm_change over time
make a bar graph of the sales vs the month
plot a bar chart of new users added over time

"""
```

```python
alt.Chart(df).mark_bar().encode(
    x="month:T",
    y="nonfarm_change",
)
```

```python
df = data.barley()
```

```python
alt.Chart(df).mark_bar().encode(
    x='year:O',
    y='sum(yield):Q',
    color='year:N',
    column='site:N'
)
```

```python
df = pd.DataFrame([['Action', 5, 'F'], 
                   ['Crime', 10, 'F'], 
                   ['Action', 3, 'M'], 
                   ['Crime', 9, 'M']], 
                  columns=['Genre', 'Rating', 'Gender'])
```

```python
alt.Chart(df).mark_bar().encode(
   column=alt.Column('Genre'),
   x='Gender',
   y='Rating',
   color='Gender'
)
```

# Histograms

```python
df = data.movies()
```

```python
df.head()
```

```python
"""
For histograms you should add bins, bar chart leave it there

Create a histogram of the count of IMDB_Rating
graph the counts of IMDB_Ratings
make a histogram of customer weight times
"""
```

```python
alt.Chart(df).mark_bar().encode(
    alt.X("IMDB_Rating", bin=True),
    y='count()',
)
```

```python
"""
Create a bar graph of the count of IMDB_Rating
make a bar graph of customer weight times
"""
```

```python
alt.Chart(df).mark_bar().encode(
    x="IMDB_Rating",
    y='count()',
)
```

# Scatter Plots

```python
df = data.cars()
```

```python
"""
create a scatter plot of Horsepower and Miles_per_Gallon
graph a scatter plot with Horsepower on the x-axis and Miles_per_Gallon on the y axis
make a scatter plot using Horsepower and Miles_per_Gallon
"""
```

```python
alt.Chart(df).mark_circle().encode(
    x='Horsepower',
    y='Miles_per_Gallon'
)
```

```python
"""
create a scatter plot of Horsepower and Miles_per_Gallon and color the points based off Origin
graph a scatter plot with Horsepower on the x-axis and Miles_per_Gallon on the y axis and change the color of the points on Origin
make a scatter plot using Horsepower and Miles_per_Gallon and mark the circles based on their Origin
"""
```

```python
alt.Chart(df).mark_circle().encode(
    x='Horsepower',
    y='Miles_per_Gallon',
    color='Origin'
)
```

```python
source = data.iris()

alt.Chart(source).mark_circle().encode(
    alt.X('sepalLength', scale=alt.Scale(zero=False)),
    alt.Y('sepalWidth', scale=alt.Scale(zero=False, padding=1)),
    color='species',
    size='petalWidth'
)
```

```python
source = data.driving()

alt.Chart(source).mark_line(point=True).encode(
    alt.X('miles', scale=alt.Scale(zero=False)),
    alt.Y('gas', scale=alt.Scale(zero=False)),
    order='year'
)
```

# Pie Charts

```python
import matplotlib.pyplot as plt
```

```python
labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes = [15, 30, 45, 10]
```

```python
fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
#ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```
